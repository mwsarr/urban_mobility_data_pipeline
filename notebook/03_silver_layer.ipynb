```python
import pyspark.sql.functions as sf
```

```python
taxi_zone_df = spark.read.table("taxi_zone_bronze")

taxi_zone_silver = taxi_zone_df.filter('service_zone != "N/A"')
taxi_zone_silver.write.mode("overwrite").format("delta").saveAsTable("taxi_zone_silver")
```

```python
yellow_taxi_trip_df = spark.read.table("yellow_taxi_trip_bronze")

yellow_taxi_silver = yellow_taxi_trip_df.select("VendorID","passenger_count","trip_distance","tpep_pickup_datetime","tpep_dropoff_datetime","PULocationID","DOLocationID","payment_type","fare_amount","tip_amount","tolls_amount","total_amount")\
        .withColumn("date", sf.to_date("tpep_pickup_datetime"))\
        .withColumn("hour", sf.date_format(("tpep_pickup_datetime"),"HH:mm:ss"))\
        .withColumnRenamed("VendorID","vendor_id")\
        .withColumnRenamed("tpep_pickup_datetime","pickup_datetime")\
        .withColumnRenamed("tpep_dropoff_datetime","dropoff_datetime")\
        .withColumnRenamed("PULocationID","pickup_location_id")\
        .withColumnRenamed("DOLocationID","dropoff_location_id")\
        .filter("trip_distance > 0")\
        .filter("fare_amount > 0")\
        

yellow_taxi_silver.write.mode("overwrite").format("delta").saveAsTable("yellow_taxi_trip_silver")
```

```python

weather_df = spark.read.table("weather_bronze")

weather_df = weather_df.withColumn("date",sf.to_date("time"))\
    .withColumn("hour",sf.date_format(("time"),"HH:mm:ss"))

weather_df.write.mode("overwrite").format("delta").saveAsTable("weather_silver")
```
